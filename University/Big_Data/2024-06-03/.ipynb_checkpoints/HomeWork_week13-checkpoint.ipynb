```python
import pandas as pd
import numpy as np
```

```python
sales = [{'account': 'jones LLC', 'Jan': 150, 'Feb' : 200 , 'Mar' : 140},
         {'account': 'Alpha Co', 'Jan': 200, 'Feb' : 210, 'Mar': 215},
         {'account': 'Blue Inc', 'Jan':50, 'Feb' : 90, 'Mar' :95}]
df = pd.DataFrame(sales)
df
```

```python
labels = ['account', 'Jan', 'Feb', 'Mar']
sales = [('jones LLC',150,200,140),('Alpha Co',200,210,215),('Blue Inc',50,90,95)]
df = pd.DataFrame.from_records(sales, columns=labels)
df
```

```python
sales = {'account': ['jones LLC', 'Alpha Co', 'Blue Inc'],'Jan':[150,200,50],'Feb':[200,210,90],'Mar':[140,215,95]}
df = pd.DataFrame.from_dict(sales)
df
```

```python
sales = [('account',['jones LLC', 'Alpha Co', 'Blue Inc']),('Jan',[150,200,50]),('Feb',[200,210,215]),('Mar',[140,215,95])]
df = pd.DataFrame.from_dict(sales)
df
```

```python
import urllib.request
import re

url = "https://finance.naver.com/item/main.nhn?code=005930"

html = urllib.request.urlopen(url)
html_contents = str(html.read())

print(html_contents)
```

```python
html = urllib.request.urlopen(url)
html_contents = str(html.read().decode('ms949'))
print(html_contents)
```

```python
stock_results = re.findall("(<dl class=\"blind\">)([\s\S]+?)(</dl>)", html_contents)
print(stock_results)
```

```python
samsung_stock = stock_results[0]
print(stock_results[0])
```

```python
samsung_index = samsung_stock[1]
print(stock_results[1])
```

```python
print(samsung_index)
```

```python
index_list = re.findall("(\<dd\>)([\s\S]+?)(\<\/dd\>)", samsung_index)
for index in index_list:
    print(index[1])
```
