```python
import boto3
import base64
import json

rekognition = boto3.client('rekognition', 'ap-northeast-2')

with open('./dog.jpg', 'rb') as cf:
    base64_image = base64.b64encode(cf.read())
    base_64_binary = base64.decodebytes(base64_image)
    print('file open!')

    response = rekognition.detect_labels(Image={'Bytes': base_64_binary})
    response

print(json.dumps(response, indent=2, ensure_ascii=False))
```

```python
import boto3
import json

rekognition = boto3.client('rekognition', region_name='ap-northeast-2')

IMG_PATH = './dog.jpg'

with open(IMG_PATH, 'rb') as f:
    img_bytes = f.read()
    print('file open!')

response = rekognition.detect_labels(
    Image={'Bytes': img_bytes},
    MaxLabels=10,
    MinConfidence=50
    )

print(json.dumps(response, indent=2, ensure_ascii=False))

with open('rekognition_response.json', 'w', encoding='utf-8') as f:
    json.dump(response, f, indent=2, ensure_ascii=False)
print('결과 저장 완료: rekognition_response.json')

for label in response['Labels']:
    print(f"- {label['Name']} ({label['Confidence']:.2f}%)")
```
