```python
from shapely.geometry import Point, LineString, Polygon
import fiona

p1 = Point(0, 0)
p2 = Point(1, 1)

line = LineString([(0, 0), (1, 2), (2, 4)])
poly = Polygon([(0, 0), (2, 0), (2, 2), (0, 2)])

print("두 점 사이의 거리:", p1.distance(p2))
print("라인 길이:", line.length)
print("폴리곤 면적:", poly.area)
print("p2가 폴리곤 내부에 있는가?", poly.contains(p2))
```
