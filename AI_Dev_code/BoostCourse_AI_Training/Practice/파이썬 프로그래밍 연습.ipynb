### [TODO] 코드 구현 1 : 행렬곱 연산
본격적으로 Numpy와 친해지기 위해서 다양한 연산을 연습해볼 예정입니다. 랜덤으로 무작위 데이터를 가진 5x3 행렬과 3x2 행렬을 numpy array로 만든 후, 행열곱 연산을 진행해봅시다. 그리고 그 결과를 출력해봅시다.


* array에 들어가는 숫자가 랜덤으로 주어지기 때문에 출력 결과는 아래와 달라질 수 있습니다.


```python
# 프로그램 실행 예시
# 출력 결과
'''[[0.15885159 0.93133205]
 [0.16377972 0.8860128 ]
 [0.14217311 0.55589135]
 [0.14671463 0.43939056]
 [0.10104267 0.33995462]] (5, 2)'''
```

```python
# 다음에 안내된 코드를 완성해보세요! 어떤 결과물이 나올까요?

import numpy as np
import random
## 코드시작 ##
# 5*3 행렬 랜덤 생성
five_three = []
for _ in range(5):
 example = []
 for i in range(3):
  x = random.random()
  example.append(x)
 five_three.append(example)
# 3*2 행렬 랜덤 생성
three_two = []
for _ in range(3):
 example = []
 for i in range(2):
  x = random.random()
  example.append(x)
 three_two.append(example)

#행렬 지정
a = np.array(five_three)
b = np.array(three_two)

#행렬 곱셈 계산
ex = np.dot(a,b)
## 코드종료 ##
print(ex, ex.shape)
```

### [TODO] 코드 구현 2 : concatenate 연산
그 다음으로는 numpy에서 자주 사용하는 concatenate 연산을 해보겠습니다. 다음 array를 사용해 axis 가 0과 1일때의 concatenate 연산을 각각 구해보세요.


*   첫번째 array : [[5,7], [9,11]]
*   두번째 array : [[2,4], [6,8]]



```python
# 프로그램 실행 예시
# 출력 결과
# [[ 5  7]
#  [ 9 11]
#  [ 2  4]
#  [ 6  8]] [[ 5  7  2  4]
#  [ 9 11  6  8]]
```

```python
# 다음에 안내된 코드를 완성해보세요! 어떤 결과물이 나올까요?
## 코드시작 ##
first_array = np.array([[5,7], [9,11]])
second_array = np.array([[2,4],[6,8]])

con = np.concatenate([first_array, second_array])
plus = np.array([
 np.concatenate([first_array[0], second_array[0]]),
 np.concatenate([first_array[1],second_array[1]])
])

## 코드종료 ##
print(con, plus)
```

### [TODO] 코드 구현 3 : Series 코드 완성하기 - 조건에 따른 목록 재구성
다음의 재고 목록을 사용해 Pandas 라이브러리의 Series 형태를 만들어보세요. 그리고 데이터의 10 이상 20 이하의 데이터만 골라 출력해봅시다.

<재고 목록>
* HDD : 19개
* SDD : 11개
* USB : 5개
* CLOUD : 97개


```python
# 프로그램 실행 예시
# 출력 결과
# If you can dream it, you can do it
```

```python
# 다음에 안내된 코드를 완성해보세요! 어떤 결과물이 나올까요?

import pandas as pd

idx = ["HDD", "SSD", "USB", "CLOUD"]
data = [19, 11, 5, 97]

## 코드시작 ##
df = pd.Series(data= data, index=idx)

serise = df[(df >= 10) & (df <= 20)]
## 코드종료 ##
print(serise)
```

### [TODO] 코드 구현 4 : dataframe 코드 완성하기 - 야채과일 가격 계산하기
다음과 같이 야채와 과일 목록이 정리된 데이터가 있습니다. 이 두 데이터를 따로 보기엔 효율성이 떨어져 1개로 합쳐 보려 합니다. 각 표에 정리된 데이터를 각각 하나의 데이터 프레임으로 생성한 뒤, 하나로 결합해보세요. 그리고 'type'을 이용해 데이터를 정렬하고, 가장 비싼 야채와 가장 비싼 과일의 가격 합을 구해보세요.

야채과일 목록.png

```python
# 프로그램 실행 예시
# 출력 결과
# 200
```

```python
# 다음에 안내된 코드를 완성해보세요! 어떤 결과물이 나올까요?

import pandas as pd

## 코드시작 ##
first_data = {'name':['cherry','mango','potato','onion'],
              'type':['fruit','fruit','vegetable','vegetable'],
              'price':[100,110,60,80]}

second_data = {'name':['pepper','carrot','banana','kiwi'],
               'type':['vegetable','vegetable','fruit','fruit'],
               'price':[50,70,90,120]}

first_frame = pd.DataFrame(first_data)
second_frame = pd.DataFrame(second_data)

mix_frame = pd.merge(first_frame,second_frame,how='outer')
mix_frame = mix_frame.sort_values('type')

sum_of_Top = max(mix_frame[mix_frame['type'] =='fruit']['price']) + max(mix_frame[mix_frame['type'] =='vegetable']['price'])
## 코드종료 ##
print(mix_frame)
print('가장 비싼 과일의 합:',sum_of_Top)
```

### [TODO] 코드 구현 5 : dataframe 코드 완성하기 - 점수 최댓값, 최솟값 출력하기

* 참여자의 이름을 인덱스로 활용해보세요.
